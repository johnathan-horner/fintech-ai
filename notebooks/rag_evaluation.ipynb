# FinTech AI — RAG Pipeline Evaluation
Evaluate retrieval quality and answer accuracy across financial queries.

```python
import json
import sys
sys.path.insert(0, '..')
from src.rag.rag_chain import query
from src.rag.vector_store import load_vector_store

# Test queries
TEST_QUERIES = [
    'What are the highest beta positions in the portfolio?',
    'Which stocks had earnings beats last quarter?',
    'What is the current Fed funds rate and how does it affect bonds?',
    'Which sectors are underperforming in the portfolio?',
    'What positions have the largest unrealized losses?',
    'Summarize NVDA recent earnings performance',
    'What is the VIX indicating about market volatility?',
    'Which analyst ratings are Sell or Strong Sell?',
]
print(f'Loaded {len(TEST_QUERIES)} test queries')
```

```python
# Test retrieval quality
vectorstore = load_vector_store()
print('Testing retrieval quality...\n')
for q in TEST_QUERIES[:3]:
    docs = vectorstore.similarity_search(q, k=4)
    print(f'Query: {q}')
    print(f'  Retrieved {len(docs)} docs:')
    for d in docs:
        print(f'    [{d.metadata.get("type", "?")}] {d.page_content[:100]}...')
    print()
```

```python
# Full RAG evaluation
results = []
for q in TEST_QUERIES:
    print(f'Testing: {q}')
    result = query(q)
    results.append({
        'question': q,
        'answer_length': len(result['answer']),
        'num_sources': len(result['sources']),
        'answer_preview': result['answer'][:200],
    })
    print(f'  Answer ({len(result["answer"])} chars, {len(result["sources"])} sources)\n')

print(f'\nEvaluation complete: {len(results)} queries tested')
```

```python
# Summary stats
import statistics
lengths = [r['answer_length'] for r in results]
sources = [r['num_sources'] for r in results]
print(f'Avg answer length: {statistics.mean(lengths):.0f} chars')
print(f'Avg sources retrieved: {statistics.mean(sources):.1f}')
print(f'Min/Max answer length: {min(lengths)} / {max(lengths)}')
```
